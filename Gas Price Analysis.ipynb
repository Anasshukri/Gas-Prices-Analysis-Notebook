## Load necessary libraries

```python
import matplotlib.pyplot as plt # Visualize data
import numpy as np # FOR creating numerical values
import pandas as pd # FOR reading csv/txt/xlsx/xls file
# import plotly.plotly as py
from math import pi
```

## Real World Examples

### Gas Price

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5), dpi=200)

plt.title('Gas Price')

# plt.plot(gas.Year, gas.USA, label="US")
# plt.plot(gas.Year, gas.Canada, label="Canada")
# plt.plot(gas.Year, gas['South Korea'], label="South Korea")

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.')
        
plt.xlabel('Year')
plt.ylabel('USD/Gallon')
# plt.xticks(gas.Year[::3])
plt.xticks(gas.Year[::3].tolist()+[2011]) # Add another point in the ticks(not in df)

country = gas.loc[:, gas.columns != 'Year'] # To Exclude 'Year' in country USED IN .legend

plt.legend(country, loc=1, bbox_to_anchor=(0.5, 0.5, 0.5, 0.5)) # Legend outside (x,y,width,height)

plt.savefig('Gas_Prices.png', dpi=400)

plt.show()
```
